```python
# Turn score rating into binary classifier (>=4.5, <4.5)
# Host_response_time into dummy variables
# Host_response_rate and host_acceptance_rate turned into decimals - remove %, /100
# room_type to dummy variables
# For cleaning the bathrooms, if bathrooms is null, extract number from bathrooms_text, else use bathrooms?
# Cleaning price - remove $
```

```python
df_variables = df[variables].copy()

# --- Clean host_response_rate and host_acceptance_rate: remove '%' and divide by 100 ---
def percent_to_float(x):
    if isinstance(x, str) and x.endswith('%'):
        return float(x.strip('%')) / 100
    return np.nan

df_variables["host_response_rate"] = df_variables["host_response_rate"].apply(percent_to_float)
df_variables["host_acceptance_rate"] = df_variables["host_acceptance_rate"].apply(percent_to_float)
df_variables
```

```python
import re
# ---  Clean bathrooms: use 'bathrooms' if not null, else extract number from 'bathrooms_text' ---
def extract_bathroom_count(row):
    if pd.notnull(row["bathrooms"]):
        return row["bathrooms"]
    elif pd.notnull(row["bathrooms_text"]):
        match = re.search(r'(\d+\.?\d*)', row["bathrooms_text"])
        if match:
            return float(match.group(1))
    return np.nan

df_variables["clean_bathrooms"] = df_variables.apply(extract_bathroom_count, axis=1)
df_variables["clean_bathrooms"]
```

```python
# --- Clean price: remove "$" and commas, convert to float ---
def clean_price(val):
    if isinstance(val, str):
        return float(val.replace("$", "").replace(",", ""))
    return np.nan

df_variables["price"] = df_variables["price"].apply(clean_price)
df_variables["price"]
```

```python
# Drop rows where the rating is missing
df_variables = df_variables[df_variables["review_scores_rating"].notnull()].copy()

# Create the binary target column
df_variables["high_rating"] = (df_variables["review_scores_rating"] >= 4.5).astype(int)

df_variables = df_variables.drop('review_scores_rating', axis=1)
df_variables
```

```python
# Group by high_rating and calculate % missing for a given column
def missing_by_rating(df, column):
    return (
        df[column].isnull()
        .groupby(df["high_rating"])
        .mean()
        .rename("percent_missing")
        .mul(100)
        .round(2)
    )

# Example for host_response_rate
missing_by_rating(df_variables, "host_response_rate")
```

```python
# Get all numeric columns (excluding the target column itself if needed)
numeric_cols = df_variables.select_dtypes(include=["number"]).columns.drop("high_rating", errors="ignore")

# Store results in a dictionary
missing_by_group = {}

# Loop through numeric columns
for col in numeric_cols:
    missing_rates = (
        df_variables[col].isnull()
        .groupby(df_variables["high_rating"])
        .mean()
        .mul(100)
        .round(2)
    )
    missing_by_group[col] = missing_rates

# Convert to DataFrame for easy viewing
missing_by_group_df = pd.DataFrame(missing_by_group).T
missing_by_group_df.columns = ["% missing (low rating)", "% missing (high rating)"]

missing_by_group_df["diff"] = (
    missing_by_group_df["% missing (low rating)"] - missing_by_group_df["% missing (high rating)"]
)
missing_by_group_df.sort_values("diff", ascending=False)
```

```python
# Create an indicator flag for missingness (since it seems to be related to high and low rating)
for col in ["price", "host_acceptance_rate", "host_response_rate"]:
    df_variables[f"{col}_missing"] = df_variables[col].isnull().astype(int)
# Impute the missing values with the median for these columns so we don't have missing values
for col in ["price", "host_acceptance_rate", "host_response_rate"]:
    median_val = df_variables[col].median()
    df_variables[col] = df_variables[col].fillna(median_val)

df_variables.info()
```

```python
# Impute bedrooms with median
df_variables["bedrooms"] = df_variables["bedrooms"].fillna(df_variables["bedrooms"].median())

# Impute clean_bathrooms with median
df_variables["clean_bathrooms"] = df_variables["clean_bathrooms"].fillna(df_variables["clean_bathrooms"].median())

# Impute host_listings_count with median
df_variables["host_listings_count"] = df_variables["host_listings_count"].fillna(df_variables["host_listings_count"].median())
```

```python
# Convert host_identity_verified to binary
df_variables["host_identity_verified"] = df_variables["host_identity_verified"].map({"t": 1, "f": 0})
df_variables["host_identity_verified"] = df_variables["host_identity_verified"].fillna(0)

# One-hot encode host_response_time, room_type, and instant_bookable
df_variables = pd.get_dummies(
    df_variables,
    columns=["host_response_time", "room_type", "instant_bookable"],
    prefix=["response_time", "room", "bookable"],
    drop_first=True,
    dummy_na=True
)
print(df_variables.isnull().sum().sort_values(ascending=False).head())
```
